```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-10, 10, 100)

# Define sigmoid function
def sigmoid(X):
    return 1 / (1 + np.exp(-X))

# Output values:
output = sigmoid(x_values)

# Plotting graph for sigmoid function
plt.plot(x_values, output)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Sigmoid Function")
plt.show()
```
