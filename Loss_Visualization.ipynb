```python
import numpy as np
import matplotlib.pyplot as plt

# Activation function (Sigmoid)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# Training Data (OR Gate)
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs = np.array([[0], [1], [1], [1]])

# Initialize Weights and Bias
np.random.seed(1)
weights = np.random.rand(2, 1)
bias = np.random.rand(1)
learning_rate = 0.5

# Tracking Loss
losses = []

# Training Process
for epoch in range(10000):
    # Forward Propagation
    input_layer = inputs
    weighted_sum = np.dot(input_layer, weights) + bias
    output = sigmoid(weighted_sum)

    # Calculate Loss
    loss = np.mean((outputs - output) ** 2)
    losses.append(loss)

    # Backpropagation
    error = outputs - output
    adjustments = error * sigmoid_derivative(output)
    weights += np.dot(input_layer.T, adjustments) * learning_rate
    bias += np.sum(adjustments) * learning_rate

# Plot Loss Graph
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()

# Prediction
test_input = np.array([[0, 1]])
pred = sigmoid(np.dot(test_input, weights) + bias)
print("Prediction for [0,1]:", pred)
```
